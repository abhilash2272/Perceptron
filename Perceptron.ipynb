```python
import numpy as np
x=np.array([1,2,3,4]) # Study hours
y=np.array([0,0,1,1]) # fail(0)/pass(1)

x,y
```

```python
# Initialization of Parameters
w=0.0
b=0.0
learning_rate=1
epochs=5

print("Initial weight",w)
print("Initial bias",b)
```

```python
# Activation Function
def step_function(z):
  return 1 if z>=0 else 0
```

```python
# Training loop
for epoch in range(epochs):
  print(f"\nEpoch {epoch+1}")
  for i in range(len(x)):
    # Compute Weighted sum
    z=x[i]*w+b

    # Compute Activation
    y_pred=step_function(z)

    # Compute Error
    error=y[i]-y_pred

    #Update Weight
    w=w+learning_rate*error*x[i]

    #Update Bias
    b=b+learning_rate*error
    print(f"Input: {x[i]}, Predicted: {y_pred}, Actual:{y[i]}, Weight: {w}, Bias: {b}")

```

```python
# Final Model parameters
print(f"Final weight: {w}")
print(f"Final bias: {b}")
```

```python
# Testing the Model
test_value=3.1
z=test_value*w+b
prediction=step_function(z)
print(f"For Study hours: {test_value}, Predicted: {prediction}")
```

Perceptron AND gate

```python
x1=np.array(
    [[0,0],
    [0,1],
    [1,0],
    [1,1]]
    )
y1=np.array([0,0,0,1])

x1,y1
```

```python
w1=np.zeros(2)
bias1=0
learning_rate1=0.1
epochs1=10

print(f"Initial weight: {w1}")
print(f"Initial bias: {bias1}")
```

```python
def step_function1(z):
  return 1 if z>=0 else 0
```

```python
for epoch in range(epochs1):
  print(f"\nEpoch {epoch+1}")
  for i in range(len(x1)):
    z1=np.dot(x1[i],w1)+bias1
    y_pred1=step_function1(z1)
    error1=y1[i]-y_pred1
    w1=w1+learning_rate1*error1*x1[i]
    bias1=bias1+learning_rate1*error1
    print(f"Prediction: {y_pred1}, Error: {error1}")
```

```python
print(f"Final weight: {w1}")
print(f"Final bias: {bias1}")
```

```python
print("Testing AND Gate using Perceptron")
for i in range(len(x1)):
  z1=np.dot(x1[i],w1)+bias1
  y_pred1=step_function1(z1)
  print(f"Input: {x1[i]}, Predicted: {y_pred1}")
```

```python
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt
w_and = w1
b_and = bias1
x_min, x_max = -0.5, 1.5
y_min, y_max = -0.5, 1.5
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                     np.linspace(y_min, y_max, 100))
Z = np.dot(np.c_[xx.ravel(), yy.ravel()], w_and) + b_and
Z = Z.reshape(xx.shape)
plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, levels=[-np.inf, 0, np.inf], colors=['#FFDDDD', '#DDFFDD'], alpha=0.8)
plt.contour(xx, yy, Z, levels=[0], colors='black', linestyles='--')
for i in range(len(x1)):
    if y1[i] == 1:
        plt.scatter(x1[i, 0], x1[i, 1], color='blue', marker='o', s=100, label='Class 1' if i == 0 else "")
    else:
        plt.scatter(x1[i, 0], x1[i, 1], color='red', marker='x', s=100, label='Class 0' if i == 0 else "")

plt.xlabel('Input 1 (x1)')
plt.ylabel('Input 2 (x2)')
plt.title('Perceptron Decision Boundary for AND Gate')
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.grid(True)
plt.legend(['Decision Boundary', 'Class 0', 'Class 1'])
plt.show()
```
